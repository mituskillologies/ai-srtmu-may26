```python
# Variables in Python
```

```python
# A variable is a name used to store data in memory.
```

```python
# Naming Conventions - Rules to declare variables in Python
```

```python
# No need to declare data type
# In Python, you don’t need to specify the type of a variable (like int, float, etc.). 
# Python automatically understands the type based on the value you assign.
```

```python
x = 10        # int
price = 99.5  # float
name = "Rahul" # string
```

```python
# Variable must start with a letter or underscore (_)
# A variable name should begin with an alphabet (a–z, A–Z) or an underscore (_). 
# Starting with anything else will cause an error.
```

```python
age = 25      # valid
```

```python
_name = "Rahul"  # valid
```

```python
1age = 25   # invalid
```

```python
# Variable cannot start with a number
# Even though numbers can be used in variable names, they cannot be placed at the beginning.
```

```python
age1 = 20   # valid
```

```python
marks2 = 90   # valid
```

```python
1marks = 90  # invalid
```

```python
# Only letters, numbers, and underscore are allowed
# Variable names cannot contain special characters like @, #, -, %, etc.
```

```python
user_name = "Rahul"   # valid
```

```python
total1 = 100          # valid
```

```python
user-name = "Rahul"   # invalid
```

```python
total@ = 100        #invalid
```

```python
# Variable names are case-sensitive
# Python treats uppercase and lowercase letters as different, 
# so variables with different cases are considered separate.
```

```python
name = "Rahul"
Name = "Amit"
```

```python
print(name)  # Rahul
```

```python
print(Name)  # Amit
```

```python
# Cannot use Python reserved keywords
# Keywords are special words that have predefined meanings in Python (like if, else, class, for). 
# These cannot be used as variable names.
```

```python
class = 10  # invalid
```

```python
value = 10   # valid
```

```python
# Use meaningful and readable names (best practice)
# Variable names should clearly describe the data they store. This makes code easier to understand.
```

```python
marks = 90       # good
```

```python
m = 90           # not clear
```

```python
student_name = "Rahul"  # better
```
