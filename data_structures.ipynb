```python
# Dictionary
```

```python
student = {
    "name": "Rahul",
    "age": 25,
    "city": "Pune"
}
```

```python
# Accessing Values
```

```python
student = {"name": "Rahul", "age": 25}

print(student["name"])
print(student.get("age"))
```

```python
# Adding Elements
```

```python
student = {"name": "Rahul"}

student["age"] = 25
student["city"] = "Pune"

print(student)
```

```python
# Updating Elements
```

```python
student = {"name": "Rahul", "age": 25}

student["age"] = 26
print(student)
```

```python
# Removing Elements
```

```python
# pop() → remove by key
```

```python
student = {"name": "Rahul", "age": 25}

student.pop("age")
print(student)
```

```python
# popitem() → removes last item
```

```python
student = {"name": "Rahul", "age": 25}

student.popitem()
print(student)
```

```python
# del → delete key or full dict
```

```python
student = {"name": "Rahul", "age": 25}

del student["age"]
print(student)
```

```python
# clear() → remove all items
```

```python
student = {"name": "Rahul", "age": 25}

student.clear()
print(student)
```

```python
# Getting Keys, Values, Items
```

```python
student = {"name": "Rahul", "age": 25}

print(student.keys())
```

```python
print(student.values())
```

```python
print(student.items())
```

```python
# Looping through Dictionary
```

```python
for key in student:
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key, value in student.items():
    print(key, value)
```

```python
# Nested Dictionary
```

```python
students = {
    "s1": {"name": "Rahul", "age": 25},
    "s2": {"name": "Amit", "age": 24}
}

print(students["s1"]["name"])
```
