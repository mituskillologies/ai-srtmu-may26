```python
import pandas as pd
```

```python
data = {
    "Study_Hours": [1, 2, 3, 4, 5, 6, 8, 9, 10],
    "Exam_Score":  [20, 25, 30, 35, 40, 45, 80, 85, 90]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Visualize Data (Before Clustering)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["Study_Hours"], df["Exam_Score"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Before Clustering")
plt.show()
```

```python
# Apply K-Means
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(df)
```

```python
df
```

```python
# Visualize Clusters
```

```python
plt.scatter(df["Study_Hours"], df["Exam_Score"], c=df["Cluster"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("K-Means Clusters")
plt.show()
```

```python
# Cluster Centers
```

```python
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

```python
# Cluster 1 → [4.5, 37.5] Meaning: Average Study Hours = 4.5 Average Exam Score = 37.5
# So this cluster represents:
# Students who study around 4–5 hours and score around 35–40 
```
