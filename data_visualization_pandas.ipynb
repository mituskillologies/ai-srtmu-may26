```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df
```

```python
# Line plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.line()
```

```python
# Bar plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.bar()
```

```python
# Horizontal bar plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.barh()
```

```python
# Box plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.box()
```

```python
# Area plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.area()
```

```python
# Density / KDE plot

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.kde()
```

```python
# Scatter plot (needs x and y)

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.scatter(x="A", y="B")
```

```python
# Hexbin plot (needs x and y)

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df.plot.hexbin(x="A", y="B", gridsize=10)
```

```python
# Pie plot (works on single column)

df = pd.DataFrame({
    "A": [50, 60, 70, 80],
    "B": [40, 55, 65, 75],
    "C": [30, 45, 55, 65],
    "D": [20, 35, 45, 55]
}, index=["Q1", "Q2", "Q3", "Q4"])

df["A"].plot.pie()
```

```python
# Pie plot

series = pd.Series([1, 2, 3, 4], index=["a", "b", "c", "d"], name="series")

series.plot.pie(figsize=(6, 6))
```

```python
# Pie plot

df = pd.DataFrame({
    "x": [10, 20, 30, 40],
    "y": [40, 30, 20, 10]
}, index=["a", "b", "c", "d"])

df.plot.pie(subplots=True, figsize=(8, 4))
```

```python
# Controlling the legend

ts = pd.date_range(start="2023-01-01", periods=10)

df = pd.DataFrame({
    "A": range(10, 110, 10),
    "B": range(20, 120, 10),
    "C": range(5, 55, 5),
    "D": range(100, 0, -10)
}, index=ts)

df = df.cumsum()

df.plot(legend=False)
```

```python
# Controlling the labels

ts = pd.date_range(start="2023-01-01", periods=10)

df = pd.DataFrame({
    "A": range(10, 110, 10),
    "B": range(20, 120, 10),
    "C": range(5, 55, 5),
    "D": range(100, 0, -10)
}, index=ts)

df = df.cumsum()

df.plot(xlabel="new x", ylabel="new y")
```

```python
# Subplots

ts = pd.date_range(start="2023-01-01", periods=10)

df = pd.DataFrame({
    "A": range(10, 110, 10),
    "B": range(20, 120, 10),
    "C": range(5, 55, 5),
    "D": range(100, 0, -10)
}, index=ts)

df = df.cumsum()

df.plot(subplots=True, figsize=(6, 6))
```

```python
# Using layout and targeting multiple axes

ts = pd.date_range(start="2023-01-01", periods=10)

df = pd.DataFrame({
    "A": range(10, 110, 10),
    "B": range(20, 120, 10),
    "C": range(5, 55, 5),
    "D": range(100, 0, -10)
}, index=ts)

df = df.cumsum()

df.plot(subplots=True, layout=(2, 3), figsize=(6, 6), sharex=False)
```
