```python
# Decision Making
```

```python
# Decision making is the process of executing different blocks of code based on conditions.
# It uses Boolean expressions (True / False)
```

```python
# if Statement
# Executes a block of code only when the condition is True
```

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

```python
# 2. if-else statement
# Runs one block if the condition is true, otherwise runs another block.
```

```python
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
```

```python
# 3. if-elif-else ladder
# Used when there are multiple conditions to check.
```

```python
x = 0

if x > 0:
    print("Positive number")
elif x < 0:
    print("Negative number")
else:
    print("Zero")
```

```python
# 4. Nested if
# An if statement inside another if.
```

```python
x = 10

if x > 5:
    if x < 20:
        print("x is between 5 and 20")
```

```python
# 5. Short-hand if (one-line if)
# Write if in a single line for simple conditions.
```

```python
x = 10

if x > 5: print("Greater than 5")
```

```python
# 6. Short-hand if-else (ternary operator)
# Write if-else in one line.
```

```python
x = 10

result = "Greater" if x > 5 else "Smaller"
print(result)
```

```python
# 7. Logical operators in decision making
# Used to combine conditions: and, or, not.
```

```python
x = 15

if x > 10 and x < 20:
    print("x is between 10 and 20")

if x < 10 or x > 12:
    print("Condition with OR")

if not x == 5:
    print("x is not equal to 5")
```

```python
# 9. Membership operators (in, not in)
# Check if a value exists in a sequence.
```

```python
list1 = [1, 2, 3, 4]

if 2 in list1:
    print("2 is present")

if 5 not in list1:
    print("5 is not present")
```

```python
# 10. Identity operators (is, is not)
# Check if two variables refer to the same object.
```

```python
a = [1, 2]
b = a

if a is b:
    print("Same object")

if a is not [1, 2]:
    print("Different objects")
```

```python
# 11. pass statement
# Used as a placeholder when no action is needed.
```

```python
x = 10

if x > 5:
    pass  # do nothing
```
