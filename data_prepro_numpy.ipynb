```python
# Array Properties
```

```python
sample_array = np.array([[1,2,3],[4,5,6]])
```

```python
print(sample_array.shape) # (rows, columns)
```

```python
# Dimension (also called an axis) refers to the number of directions needed to find a specific piece of data.

# 0-D Array (Scalar): A single number (e.g., 5). It has no dimensions.
# 1-D Array (Vector): A simple list of numbers (e.g., [1, 2, 3]). You only need one index to find a value.
# 2-D Array (Matrix): A table with rows and columns. You need two indices (row, column) to find a value.
# 3-D Array (Tensor): A stack of tables (like a book of spreadsheets). You need three indices (page, row, column) to find a value. 
```

```python
print(sample_array.ndim) # dimensions
```

```python
print(sample_array.size) # total elements
```

```python
print(sample_array.dtype) # data type
```

```python
# Special Arrays
```

```python
np.zeros((2,3))   # all zer
```

```python
np.ones((2,3))    # all ones
```

```python
np.arange(0,10,2) # range
```

```python
# np.arange(start, stop, step)
# Focuses on the interval between numbers.
# Result: [0, 2, 4, 6, 8]
# Behavior: Similar to Python's range. It starts at 0 and adds 2 repeatedly. It excludes the stop value (10).
# Best Use: When you know exactly how far apart you want your numbers to be.
```

```python
np.linspace(0,1,5) # equally spaced values
```

```python
# np.linspace(start, stop, num)
# Focuses on the total count of numbers.
# Result: [0. , 0.25, 0.5 , 0.75, 1. ]
# Behavior: "Linear Space." It squeezes exactly 5 numbers between 0 and 1. It includes the stop value (1) by default.
# Best Use: When you need a specific number of points to create a smooth line or graph.
```

```python
# Axis Concept
```

```python
# In NumPy, Axis is just another word for Dimension. 
# You use axes to tell NumPy which direction to perform a calculation (like summing numbers).
```

```python
sample_array = np.array([[1,2],[3,4]])
```

```python
print(np.sum(sample_array, axis=0))  # column-wise
```

```python
print(np.sum(sample_array, axis=1))  # row-wise
```

```python
# NaN & Missing Values
```

```python
# In NumPy, NaN stands for "Not a Number". 
# It is a special floating-point value used to represent missing or undefined data.
```

```python
sample_array = np.array([1, np.nan, 3])
```

```python
print(np.isnan(sample_array))
```

```python
print(np.nanmean(sample_array))
```
